### Визуализация занимает важную часть в анализе данных. 

Представляя информацию в графическом виде, вы **облегчаете процесс восприятия данных**, что дает возможность выделять дополнительные закономерности, оценивать соотношения величин, быстрее доносить ключевые аспекты в данных.

Начнем с небольшой "памятки", о которой всегда нужно помнить при создании любых графиков.

<div class="alert alert-danger">
<h3>
    <p style="color:black; text-align:left">
        Запрещается делать при работе с визуалом
    </p>
</h3>
</div>

1. Оставлять график без названия. По рисунку нельзя понять, какие данные и явления он описывает.
2. Оставлять график без подписи осей.
3. Не указывать единицы измерения.
4. Оставлять размер рисунка дефолтным (по дефолту размер - маленький). Все подписи, описания, легенда, название должны свободно читаться.
5. Пытаться уместить всю информацию на одном графике. Каждому гипотезе/выводу - свой график!
6. Нагромождать график без необходимости фишками, найденными в ответах mail.ru

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/karpovmax/datasets_python_teaching/master/german_credit.csv')
data.head()
```

## Bar chart - Столбчатая диаграмма

**Столбчатая диаграмма** отображает сравнение нескольких дискретных категорий. 

Одна её ось показывает сравниваемые категории, другая — измеримую величину. Иногда столбчатые диаграммы отображают несколько величин для каждой сравниваемой категории. ([_подробнее_](https://chartio.com/learn/charts/bar-chart-complete-guide/))

```python
sns.countplot(x=data.purpose); # ~ .plt(kind='bar')
```

Довольно неудобный график.

Рассмотрим по пунктам, как его можно улучшить.

```python
plt.figure(figsize=(18, 10))
sns.countplot(x=data.purpose);
```

* Первый способ: отдельно по каждой категории

* Второй способ: метод наложения

```python
plt.figure(figsize=(18, 10))

# задний слой
backgound = sns.countplot(x=data.purpose, # инициализация данных
                          color='red') # цвет графика

# верхний слой
upper = sns.countplot(x=data[data.credit_risk == 'good'].purpose, # отделяем из всех данных только те, которые нас интересуют
                     color = 'green'); # задаем цвет
```

## Scatterplot - Точечная диаграмма

Точечная диаграмма используется, чтобы определить, имеют ли две переменные взаимосвязь или корреляцию. ([_подробнее_](https://chartio.com/learn/charts/what-is-a-scatter-plot/))

```python
plt.figure(figsize=(14, 10))
plot = sns.scatterplot(x='amount', y='duration', hue='credit_risk', data=data,
                       palette=['green', 'red'])
plot.set(title='Duration depending on amount by credit risk',
         xlabel='Credit amount', 
         ylabel='Duration');
```

1. Изменим размер точек
2. Изменим легенду

```python
# 1

plt.figure(figsize=(14, 10))
plot = sns.scatterplot(x='amount', y='duration', hue='credit_risk', data=data,
                       palette=['green', 'red'], 
                       s=50)  # s - отвечает за size
plot.set(title='Duration depending on amount by credit risk',
         xlabel='Credit amount', 
         ylabel='Duration');
```

```python
# 2

plt.figure(figsize=(14, 10))
plot = sns.scatterplot(x='amount', y='duration', hue='credit_risk', data=data,
                       palette=['green', 'red'], 
                       s=50)  # s - отвечает за size
plot.set(title='Duration depending on amount by credit risk',
         xlabel='Credit amount', 
         ylabel='Duration')

handles, labels  =  plot.get_legend_handles_labels() # если не вдаваться в подробности, смысл функции в сборе координат
plot.legend(handles, ['Good credit risk', 'Bad credit risk']); # изменение легенды\
```

## Boxplot

Боксплот, или по-другому, ящик с усами - график, компактно и в удобной форме показывающий:
* Медиану (или, если нужно, среднее);
* Нижний и верхний квартили (25% и 75% соответственно);

([_подробнее_](https://chartio.com/learn/charts/box-plot-complete-guide/))

### Какую зависимость можно получить по графику?

## Violin plot

**Violin plot** представляет собой гибрид ящика с усами (боксплот) и графика плотности, который показывает пики в данных.

_Скрипка_ используется для визуализации распределения числовых данных. В отличие от боксплота, который может отображать только сводную статистику, графики скрипки отображают сводную статистику и плотность каждой переменной. ([_подробнее_](https://chartio.com/learn/charts/violin-plot-complete-guide/))

```python
sns.catplot(x='savings', y='age', hue='credit_risk', kind="violin", data=data);
```

Страшный и некрасивый график: изменим его. 

Помимо ранее применяемой функции `plt.figure(figzise=(...))`, некоторые графики позволяют изменять размер посредством параметров:

```python
sns.catplot(x='savings', y='age', hue='credit_risk', kind="violin", data=data, 
            height=5, aspect=2.3);
```

Соединим полученные "скрипки":

```python
sns.catplot(x='savings', y='age', hue='credit_risk', kind="violin", data=data, 
               height=5, aspect=2.3,
               split=True);
# Несмотря на то, что мы СОЕДИНЯЕМ, функция называется split - разделять :)
```

Изменим цвет и подпишем график:

```python
g = sns.catplot(x='savings', y='age', hue='credit_risk', kind="violin", data=data,
                height=5, aspect=2.3,
                palette=['green', 'red'], # меняем цветовую палитру
                split=True) 

# Подписываем график
g.set(xlabel='Savings categories', 
      ylabel='Age', 
      title='Clients distribution by savings, age and credit risk');
```

### Что показывает этот график?

## Heatmap - Тепловая карта

Тепловая карта отображает взаимосвязи **числовых** переменных в виде сетки цветных квадратов.

Переменные оси разделены на диапазоны и цвет каждой ячейки указывает значение основной переменной в соответствующем диапазоне ячеек. ([_подробнее_](https://chartio.com/learn/charts/heatmap-complete-guide/))

Чтобы исследовать взаимосвязь числовым переменных на статус кредитного риска, необходимо `credit_risk` перевести в числовой формат:

```python
data['credit_risk'] = data['credit_risk'].apply(lambda x: 1 if x == 'good' else 0)
```

На деле, большую часть представленных признаков (кроме категориальных) из таблицы можно преобразовать в числовые показатели, однако для примера оставим 4 показателя.

```python
cor
```

```python
plt.figure(figsize=(10, 8))

cor = data.corr() #считает корреляции для количественных признаков (фичей)
sns.heatmap(cor);
```

Сделаем тепловую карту более понятной

```python
plt.figure(figsize=(10, 8))

cor = data.corr()
sns.heatmap(cor, 
            annot=True, # отвечает за подписи значений внутри квадратов
            vmin=-1, vmax=1, center= 0,  # меняет масштаб разброса
            cmap= 'coolwarm', # меняет цветовую палитру
            fmt='.1g'); # меняет количество знаков после запятой
```

### Какие выводы можно сделать?

### Лайфхак

Тепловой картой можно пользоваться для визуализации пропусков в датафрейме!

Рисование пустых значений в датафрейме:

```python
import random
```

```python
plt.figure(figsize=(10, 8))

# Для примера заменим некоторые строки на пустые
for i in range(50):
    data.loc[random.randint(1, 1000)] = None
    
# Тогда график будет выглядеть:
sns.heatmap(data.isnull(), # инициализация данных
            yticklabels=False, # убрать подписи оси "У"
            cbar=False, # убрать цветовую палитру сбоку
            cmap='viridis'); # цветовая палитра
```

## Несколько графиков

```python
fig, ax = plt.subplots(2, 2, figsize=(13,10))

x = np.linspace(-np.pi, np.pi, 300)
y = np.sin(x)

# Первый график (line график)
ax[0,0].plot(x, y)
ax[0,0].set_title('синус')
ax[0][0].grid() # показать сетку

# Второй график (количественный)
sns.countplot(x=data.housing, # тип графика
              ax=ax[0,1]) # расположение

# Третий график (график плотности)
sns.kdeplot(x=data.age,
            shade=True,
            ax=ax[1,0])
ax[1][0].grid()

# Четвертый график (диаграмма рассеяния)
gr = sns.scatterplot(x='amount', y='age', hue='credit_risk', data=data, palette=['green', 'red'],
                ax=ax[1,1]);
handles, labels  =  gr.get_legend_handles_labels() # если не вдаваться в подробности, смысл функции в сборе координат
gr.legend(handles, ['Good credit risk', 'Bad credit risk']);
```
